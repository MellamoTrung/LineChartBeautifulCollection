```python
import matplotlib.pyplot as plt
import numpy as np

# Korean Air color palette
colors = {
    'sky_blue': '#009ADE',
    'red': '#C8102E',
    'dark_blue': '#00205B',
    'white': '#FFFFFF',
    'gray': '#A7A8AA'
}

# Data
x = np.array([1, 4, 6])
y = np.array([8, 19, 22])

# Create the figure and axis with a white background
fig, ax = plt.subplots()

# Customize the line plot with Korean Air colors
ax.plot(x, y, color=colors['sky_blue'], linewidth=3, marker='o', markersize=10, markerfacecolor=colors['red'], markeredgewidth=2)

# Add data labels
for i in range(len(x)):
    ax.annotate(f'{y[i]}', (x[i], y[i]), textcoords="offset points", xytext=(0, 10), ha='center', fontsize=12, color=colors['dark_blue'])

# Set the background color to white
ax.set_facecolor(colors['white'])
fig.patch.set_facecolor(colors['white'])

# Remove top and right spines (for a cleaner look)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Thicken the remaining spines and change their color to dark blue
ax.spines['left'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)
ax.spines['left'].set_color(colors['dark_blue'])
ax.spines['bottom'].set_color(colors['dark_blue'])

# Customize ticks
ax.tick_params(axis='both', which='major', labelsize=12, color=colors['dark_blue'], width=1.5)

# Remove gridlines for a cleaner look
ax.grid(False)

# Set titles with custom font, size, and padding
ax.set_title('Line Plot with Korean Air Color Scheme', fontsize=16, fontweight='bold', color=colors['dark_blue'], pad=20)
ax.set_xlabel('X Axis', fontsize=14, fontweight='medium', color=colors['dark_blue'])
ax.set_ylabel('Y Axis', fontsize=14, fontweight='medium', color=colors['dark_blue'])

# Set tight layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
