```python
import matplotlib.pyplot as plt
import numpy as np

# Asiana Airlines color palette
colors = {
    'red': '#B71234',
    'yellow': '#FFB612',
    'gray': '#CCCCCC',
    'dark_gray': '#666666',
    'light_gray': '#E6E6E6',
    'white': '#FFFFFF'
}

# Data
x = np.array([0, 4, 6])
y = np.array([0, 19, 22])

# Create the figure and axis
fig, ax = plt.subplots()

# Customize the line plot with Asiana Airlines colors
ax.plot(x, y, color=colors['red'], linewidth=3, marker='o', markersize=10, markerfacecolor=colors['yellow'], markeredgewidth=2)

# Add data labels
for i in range(len(x)):
    ax.annotate(f'{y[i]}', (x[i], y[i]), textcoords="offset points", xytext=(0, 10), ha='center', fontsize=12, color=colors['white'])

# Create a gradient background using light gray and white
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.patches as patches

gradient = LinearSegmentedColormap.from_list('gradient', [colors['light_gray'], colors['gray']])

# Apply the gradient as the background
ax.set_facecolor('none')
fig.patch.set_facecolor('none')

gradient_patch = patches.Rectangle((0, 0), 1, 1, transform=ax.transAxes, color='none', zorder=-1)
ax.add_patch(gradient_patch)
ax.patch.set_facecolor(gradient(0))
fig.patch.set_facecolor(gradient(1))

# Remove top and right spines (for a cleaner look)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Thicken the remaining spines and change their color
ax.spines['left'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)
ax.spines['left'].set_color(colors['dark_gray'])
ax.spines['bottom'].set_color(colors['dark_gray'])

# Customize ticks
ax.tick_params(axis='both', which='major', labelsize=12, color=colors['dark_gray'], width=1.5)

# Remove gridlines for a cleaner look
ax.grid(False)

# Set titles with custom font, size, and padding
ax.set_title('Line Plot with Asiana Airlines Color Scheme', fontsize=16, fontweight='bold', color=colors['dark_gray'], pad=20)
ax.set_xlabel('X Axis', fontsize=14, fontweight='medium', color=colors['dark_gray'])
ax.set_ylabel('Y Axis', fontsize=14, fontweight='medium', color=colors['dark_gray'])

# Set tight layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
